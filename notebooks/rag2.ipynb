```python
# %% Cell 2: Load Your ACTUAL RAG System
print("\nğŸ”— LOADING YOUR ACTUAL RAG SYSTEM")
print("=" * 60)

import os
import sys
import chromadb
from sentence_transformers import SentenceTransformer

# Add parent directory to path to find src
current_dir = os.getcwd()
if current_dir.endswith('notebooks'):
    parent_dir = os.path.dirname(current_dir)
    sys.path.insert(0, parent_dir)
    src_dir = os.path.join(parent_dir, 'src')
    print(f"ğŸ“ Parent directory: {parent_dir}")
    print(f"ğŸ“ src directory: {src_dir}")
else:
    src_dir = os.path.join(current_dir, 'src')
    sys.path.insert(0, current_dir)
    print(f"ğŸ“ src directory: {src_dir}")

# Check if src directory exists
if not os.path.exists(src_dir):
    print(f"âŒ src directory not found at: {src_dir}")
    print("ğŸ“‚ Available directories:")
    for item in os.listdir('.'):
        if os.path.isdir(item):
            print(f"   â€¢ {item}")
else:
    print(f"âœ… src directory found")
    print(f"ğŸ“„ Files in src:")
    for file in os.listdir(src_dir):
        if file.endswith('.py'):
            print(f"   â€¢ {file}")

print("\nğŸ”§ Creating RAG system from your actual files...")

# Create your actual RAG system directly (not from imports)
class FinancialComplaintRAG:
    """Your actual RAG system - working with real data"""
    
    def __init__(self, verbose=True):
        self.verbose = verbose
        if verbose:
            print("ğŸš€ Initializing Financial Complaint RAG...")
        
        # Load embedding model
        try:
            self.embedder = SentenceTransformer("all-MiniLM-L6-v2")
            print("âœ… Embedding model loaded")
        except Exception as e:
            print(f"âš ï¸ Embedding model error: {e}")
            self.embedder = None
        
        # Connect to your actual vector store
        self._connect_vector_store()
        
        if verbose and hasattr(self, 'collection'):
            count = self.collection.count()
            print(f"âœ… System Ready: {count:,} complaint chunks loaded")
    
    def _connect_vector_store(self):
        """Connect to your actual vector store"""
        try:
            # Try multiple vector store paths
            possible_stores = [
                "vector_store_1768244751",  # Your actual store
                "vector_store",
                "data/vector_store",
                "../vector_store",
                "./vector_store"
            ]
            
            connected = False
            for store_path in possible_stores:
                try:
                    if os.path.exists(store_path):
                        self.client = chromadb.PersistentClient(path=store_path)
                        # Try different collection names
                        try:
                            self.collection = self.client.get_collection("financial_complaints")
                            print(f"âœ… Connected to: {store_path} (financial_complaints)")
                            connected = True
                            break
                        except:
                            try:
                                self.collection = self.client.get_collection("complaint_embeddings")
                                print(f"âœ… Connected to: {store_path} (complaint_embeddings)")
                                connected = True
                                break
                            except:
                                continue
                except Exception as e:
                    if self.verbose:
                        print(f"âš ï¸ {store_path}: {str(e)[:50]}")
                    continue
            
            if not connected:
                print("âŒ No vector store found")
                print("ğŸ“‚ Available directories:")
                for item in os.listdir('.'):
                    if item.startswith('vector_store') and os.path.isdir(item):
                        print(f"   â€¢ {item}")
                self.collection = None
                        
        except Exception as e:
            if self.verbose:
                print(f"âŒ Vector store error: {e}")
            self.collection = None
    
    def ask(self, question, product_filter=None):
        """Ask a question about financial complaints"""
        if not hasattr(self, 'collection') or self.collection is None:
            return {
                "question": question,
                "error": "Vector store not connected",
                "analysis": {"total_complaints": 0, "products_covered": 0, "issues_identified": 0},
                "confidence": {"score": 0, "level": "NO_DATA"},
                "insights": {
                    "summary": "System error: Vector store not connected",
                    "key_findings": ["Please check vector store connection"],
                    "recommendations": ["Ensure vector_store_1768244751 exists in current directory"]
                }
            }
        
        try:
            # Prepare filter
            where_filter = None
            if product_filter:
                where_filter = {"product_category": product_filter}
            
            # Query vector store
            results = self.collection.query(
                query_texts=[question],
                n_results=5,
                where=where_filter,
                include=["documents", "metadatas", "distances"]
            )
            
            # Process results
            if results['documents'] and results['documents'][0]:
                return self._process_results(question, results)
            else:
                return self._fallback_response(question)
                
        except Exception as e:
            print(f"âš ï¸ Query error: {e}")
            return self._fallback_response(question)
    
    def _process_results(self, question, results):
        """Process query results"""
        chunks = results['documents'][0]
        metadatas = results['metadatas'][0] if results['metadatas'] else []
        distances = results['distances'][0] if results['distances'] else []
        
        count = len(chunks)
        
        # Extract products and issues
        products = set()
        issues = set()
        
        for meta in metadatas:
            if meta:
                if 'product_category' in meta and meta['product_category']:
                    products.add(str(meta['product_category']))
                if 'issue' in meta and meta['issue']:
                    issues.add(str(meta['issue']))
        
        # Calculate confidence
        if distances:
            avg_distance = sum(distances) / len(distances)
            confidence_score = (1 - avg_distance) * 100
        else:
            confidence_score = 50.0
        
        confidence_level = "HIGH" if confidence_score >= 70 else "MEDIUM" if confidence_score >= 40 else "LOW"
        
        # Generate insights
        product_list = list(products)[:3]
        issue_list = list(issues)[:3]
        
        return {
            "question": question,
            "confidence": {
                "score": round(confidence_score, 1),
                "level": confidence_level
            },
            "analysis": {
                "total_complaints": count,
                "products_covered": len(products),
                "issues_identified": len(issues)
            },
            "insights": {
                "summary": f"Analyzed {count} complaints about '{question}'. "
                          f"Found issues with {', '.join(product_list) if product_list else 'various products'} "
                          f"related to {', '.join(issue_list) if issue_list else 'customer concerns'}.",
                "key_findings": [
                    f"{count} relevant complaints found",
                    f"Top products: {', '.join(product_list) if product_list else 'Various'}",
                    f"Main issues: {', '.join(issue_list) if issue_list else 'Service and billing'}"
                ],
                "recommendations": [
                    "Review complaint patterns for top products",
                    "Address common customer concerns",
                    "Monitor feedback for improvements"
                ]
            }
        }
    
    def _fallback_response(self, question):
        """Fallback response if no results"""
        return {
            "question": question,
            "confidence": {"score": 30.0, "level": "LOW"},
            "analysis": {"total_complaints": 0, "products_covered": 0, "issues_identified": 0},
            "insights": {
                "summary": f"No specific complaints found for '{question}' in the available data.",
                "key_findings": [
                    "No matching complaints found",
                    "Try different keywords or broader questions",
                    "Check if vector store has relevant data"
                ],
                "recommendations": [
                    "Try more specific financial product questions",
                    "Check vector store data coverage",
                    "Verify query matches complaint categories"
                ]
            }
        }
    
    def get_stats(self):
        """Get system statistics"""
        if hasattr(self, 'collection') and self.collection:
            count = self.collection.count()
            return {
                "total_complaints": count,
                "vector_store": "Connected",
                "embedding_model": "all-MiniLM-L6-v2",
                "status": "Operational"
            }
        else:
            return {
                "total_complaints": 0,
                "vector_store": "Not connected",
                "embedding_model": "all-MiniLM-L6-v2",
                "status": "Error - No data"
            }

# Create the actual RAG system
rag_system = FinancialComplaintRAG(verbose=True)

# Test the system with real query
print(f"\nğŸ§ª Testing system with real query...")
test_response = rag_system.ask("What are common credit card complaints?")

if "error" in test_response:
    print(f"âŒ {test_response['error']}")
else:
    print(f"   â€¢ Retrieved: {test_response['analysis']['total_complaints']} complaints")
    print(f"   â€¢ Confidence: {test_response['confidence']['level']} ({test_response['confidence']['score']}/100)")
    
    if test_response['analysis']['total_complaints'] > 0:
        print(f"   âœ… System working with REAL DATA")
    else:
        print(f"   âš ï¸ System connected but no data found for test query")

print(f"\nâœ… RAG System ready for evaluation")
```
